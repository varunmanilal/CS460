```python
#1. Using sklearn
#Libraries imported.
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Taking 100 random points bw 1 and 0
x = random.rand(100)
c = random.rand(100)
#Plotting points
plt.scatter(x,4*x+c,label='Data')
y = 4*x+c
plt.xlabel("X Value")
plt.ylabel("Y Value")

#Changing dimension of X array
x = np.array(x).reshape((-1,1))
#print(x)
#print(y)
#using the Linear Regression Function to fit
model = LinearRegression()
model.fit(x,y)
#Print slope and intercept
print('The intercept: ',model.intercept_ )
p = model.intercept_
print('The slope: ',model.coef_[0])
q = model.coef_[0]
#Show plot with slope
plt.plot(x,q*x+p,'g')
plt.show()
```

```python
#2. Batch gradient descent
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
#from sklearn.linear_model import LinearRegression

#taking 100 points and plotting
n=100
x = random.rand(n)
c = random.rand(n)
plt.scatter(x,4*x+c,label='Data')
y = 4*x+c
plt.xlabel("X Value")
plt.ylabel("Y Value")


# Random values for slope and intercept between 0 and 10.
m = random.random()*10
b = random.random()*10
#plt.plot(x,x*m+c,'r')
#Value of alpha-learning rate
alpha = 0.3 
j = 0
#Taking 500 iterations for convergence
for i in range(500):
    j = 0
    for l in range(n):
        #calculating loss fn
        j += 1/n*pow(x[l]*m+b-y[l],2)
    delm = 0
    delb = 0
    for k in range(n):
        #error in slope and intercept
        delm += 1/n*2*(m*x[k]+b-y[k])*x[k]
        delb += 1/n*2*(m*x[k]+b-y[k])
    #updating slope and intercept
    m = m - alpha*delm
    b = b - alpha*delb
#Printing final loss function, slope and intercept
print("The loss function :",j)
print("The slope: ",m)
print("The intercept: ",b)
#Plotting it
plt.plot(x,x*m+b,'g')
plt.show()
```

```python
#3. mini-batch GD
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
#from sklearn.linear_model import LinearRegression

n=100
x = random.rand(n)
c = random.rand(n)

plt.scatter(x,4*x+c,label='Data')
y = 4*x+c
plt.xlabel("X Value")
plt.ylabel("Y Value")
# Dividing into mini-batches each having 10 numbers
x.resize(10,10)
y.resize(10,10)

# Random values for slope and intercept between 0 and 10.
m = random.random()*10
b = random.random()*10
#value of alpha
alpha = 0.3 
j = 0
#Taking 500 ierations
for i in range(500):
    #Moving through all 10 batches
    for z in range(10):
        j = 0
        for l in range(10):
            #Loss fn
            j += 1/10*pow(x[z][l]*m+b-y[z][l],2)
        delm = 0
        delb = 0
        for k in range(10):
            # delta slope and delta intercept
            delm += 1/10*2*(m*x[z][k]+b-y[z][k])*x[z][k]
            delb += 1/10*2*(m*x[z][k]+b-y[z][k])
        # shifting slope and intercept
        m = m - alpha*delm
        b = b - alpha*delb

print("The loss function :",j)
print("The slope: ",m)
print("The intercept: ",b)
plt.plot(x,x*m+b,'g')
plt.show()
```

```python
#4. SGD
import numpy as np
from numpy import random
import matplotlib.pyplot as plt

n=100
x = random.rand(n)
c = random.rand(n)

plt.scatter(x,4*x+c,label='Data')
y = 4*x+c
plt.xlabel("X Value")
plt.ylabel("Y Value")

# Random values for slope and intercept between 0 and 10.
m = random.random()*10
b = random.random()*10
#Value of alpha
alpha = 0.1 
j = 0
#500 iterations over every point
for i in range(500):
    # Covering all 100 points
    for l in range(n):
        j = 0
        #loss function
        j += pow(x[l]*m+b-y[l],2)
        delm = 0
        delb = 0
        delm += 2*(m*x[l]+b-y[l])*x[l]
        delb += 2*(m*x[l]+b-y[l])
        m = m - alpha*delm
        b = b - alpha*delb
print("The loss function :",j)
print("The slope: ",m)
print("The intercept: ",b)
plt.plot(x,x*m+b,'g')
plt.show()

        
```
