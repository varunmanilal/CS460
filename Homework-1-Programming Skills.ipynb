```python
import numpy as np
import matplotlib.pyplot as plt

mean = [0,0]
covariance = [[1,0],[0,1]]

(x,y) = np.random.multivariate_normal(mean,covariance,100).T

plt.plot(x,y,'o')
plt.title("Gaussian distribution for mean={} and covariance={}".format(mean,covariance))
plt.show()


```

```python
mean = [-1,1]

(x,y) = np.random.multivariate_normal(mean,covariance,100).T

plt.plot(x,y,'o',color = 'g')
plt.title("Gaussian distribution for mean={} and covariance={}".format(mean,covariance))
plt.show()
```

```python
mean = [0,0]
covariance = [[2,0],[0,2]]
(x,y) = np.random.multivariate_normal(mean,covariance,100).T

plt.plot(x,y,'o',color = 'black')
plt.title("Gaussian distribution for mean={} and covariance={}".format(mean,covariance))
plt.show()
```
